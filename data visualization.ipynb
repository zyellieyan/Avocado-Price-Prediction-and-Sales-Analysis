```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
largedata = pd.read_csv('large_region.csv')
smalldata = pd.read_csv('small_region.csv')
```

```python
largedata.head()
```

```python
groupregion_large = largedata.groupby(by='region').mean()
groupregion_large
```

```python
groupregion_small = smalldata.groupby(by='region').mean()
groupregion_small
```

```python
plt.figure(figsize=(16,8))
plt.plot(groupregion_large['AveragePrice'])
```

```python
plt.figure(figsize=(25,8))
plt.plot(groupregion_small['AveragePrice'])
plt.xticks(rotation=60,fontsize=10)
```

```python
datechart = largedata.groupby(by = 'Date').mean()
datechart.head()
datechart
```

```python
plt.figure(figsize=(25,6))
plt.plot(datechart.index,datechart['AveragePrice'])
```
