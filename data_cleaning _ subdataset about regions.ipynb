```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('avocado.csv')
data.info()
```

```python
data.head()
```

```python
data = data.drop('Unnamed: 0',axis=1)
data.head()
```

```python
data.var()
```

```python
data['type'] = 1*  data['type'] == 'organic'
```

```python
data['type'] = 1* data['type']
```

```python
data['type'].mean()
```

```python
region_value = data.region.value_counts()
region_value = pd.DataFrame(region_value)
regionlist= list(region_value.index)
len(regionlist)
```

```python
regionlist
```

```python
largelist = ['California','GreatLakes','Midsouth','Northeast','SouthCarolina','SouthCentral','Southeast','West',
             'TotalUS','WestTexNewMexico','Plains','NorthernNewEngland']
len(largelist)
```

```python
data_largeregion = data[data['region'].isin(largelist)]
```

```python
data_largeregion.region.value_counts()
```

```python
smalllist = region_value.index.drop(largelist)
```

```python
smalllist = list(smalllist)
len(smalllist)
```

```python
data_smallregion = data[data['region'].isin(smalllist)]
data_smallregion.region.value_counts()
```

```python
np.count_nonzero(data_smallregion.region.unique())
```

```python
data_largeregion.to_csv('large_region.csv',index=None)
data_smallregion.to_csv('small_region.csv',index=None)
```
